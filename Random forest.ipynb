```python
# One-hot encode the data using pandas get_dummies
temp= pd.get_dummies(temp)
# Display the first 5 rows of the last 12 columns
temp.iloc[:,5:].head(12)
```

Here we do one-hot encode in the data so the string data is conveted into numerical data by using get_dummies

```python
import numpy as np
labels = np.array(temp['actual'])
temp= temp.drop('actual', axis = 1)
temp_list = list(temp.columns)
temp= np.array(temp)
```

```python
from sklearn.model_selection import train_test_split
train_temp, test_temp, train_labels, test_labels = train_test_split(temp, labels, test_size = 0.25, random_state = 0)
```

```python
print('Training temp :', train_temp.shape)
print('Training labels :', train_labels.shape)
print('Testing temp :', test_temp.shape)
print('Testing labels :', test_labels.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 1000, random_state =0)
rf.fit(train_temp, train_labels);
```

```python
predictions = rf.predict(test_temp)
errors = abs(predictions - test_labels)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
mape = 100 * (errors / test_labels)
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
