```python
import pandas as pd
import numpy as np
```

```python
Companies_Df=pd.read_csv("companies.txt",sep="\t",encoding="ANSI")
Companies_Df
```

```python
Rounds2=pd.read_csv("rounds2.csv",encoding="ANSI")
Rounds2.head()
```

```python
Rounds2.company_permalink=Rounds2.company_permalink.str.lower()
Rounds2.company_permalink
```

```python
Companies_Df.permalink=Companies_Df.permalink.str.lower()
Companies_Df.permalink
```

```python
Companies_Df.permalink.nunique()
```

```python
Rounds2.company_permalink.nunique()
```

```python
Rounds2=Rounds2.rename({"company_permalink":"permalink"}, axis='columns')
Rounds2.head()
```

```python
master_df=pd.merge(Rounds2,Companies_Df,how='inner',on='permalink')
```

```python
master_df.head()
```

```python
master_df.info()
```

```python
round(100*(master_df.isnull().sum()/len(master_df.index)), 2)
```

```python
#Remove the columns founded_at, funding_round_code, homepage_url  

master_df= master_df.drop('funding_round_code',axis=1)
master_df= master_df.drop('founded_at',axis=1)
master_df= master_df.drop('homepage_url',axis=1)
                      
```

```python
round(100*(master_df.isnull().sum()/len(master_df.index)), 2)
```

```python
#remove the rows where raised_amount_usd has null values.
master_df = master_df[~np.isnan(master_df['raised_amount_usd'])]
```

```python
round(100*(master_df.isnull().sum()/len(master_df.index)), 2)
```

```python
#use dropna function to drop remaining missing values inside each row.
master_df.dropna(axis=0,how="any",inplace=True)
```

```python
round(100*(master_df.isnull().sum()/len(master_df.index)), 2)
```

```python
master_df.shape
```

```python
#Above all Only venture falls in the category to invest between 5-15 million usd.
#So we have to perform analysis further only for venture type of fundings.

master_df=master_df[master_df.funding_round_type=="venture"]
master_df.shape
```

```python
#To choose the top 3 english speaking countries to invest in where maximum people have invested in
master_df.groupby("country_code").raised_amount_usd.sum().sort_values(ascending=False)
```

```python
#Sector wise analysis
#Spliting into one primary sector
master_df['primary_sector']=master_df.category_list.map(lambda x: str(x).split('|')[0])
```

```python
master_df.head()
```

```python
#Importing mapping csv files
mapping=pd.read_csv("mapping.csv",encoding="ANSI")
mapping['category_list'] = mapping['category_list'].str.replace('0','na')
mapping
```

```python
#Offset which consists of only category_list and the main_sector
offset_mapping=pd.DataFrame() 
```

```python
#changing the index to apply argmax
mapping.set_index("category_list",inplace=True)
```

```python
#Creating a new column Main_Sector which has the main_sector of the category
offset_mapping['Main_Sector']=mapping.idxmax(axis=1)
offset_mapping
```

```python
#Reset the index back to mapping
offset_mapping.reset_index(level=0, inplace=True)
offset_mapping
```

```python
#Renaming the column to primary_sector to merge the dataframes
offset_mapping=offset_mapping.rename({"category_list":"primary_sector"}, axis='columns')
offset_mapping
```

```python
master_df=pd.merge(master_df, offset_mapping, how='inner', on=['primary_sector'])
```
